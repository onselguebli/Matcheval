```python
import requests
import pandas as pd

# Param√®tres de recherche
mot_cle = "data"
base_url = "https://www.keejob.com/api/v1/job/"
headers = {
    "User-Agent": "Mozilla/5.0"
}

offres = []
for page in range(1, 6):  # Change le range si tu veux plus de pages
    params = {
        "keywords": mot_cle,
        "page": page
    }

    response = requests.get(base_url, params=params, headers=headers)
    
    if response.status_code != 200:
        print(f"Erreur sur la page {page} (code {response.status_code})")
        break

    data = response.json()

    if not data["results"]:
        break  # Stop si pas d'autres r√©sultats

    for offre in data["results"]:
        offres.append({
            "Titre": offre["title"],
            "Entreprise": offre["company"]["name"] if offre.get("company") else "N/A",
            "Lieu": offre["location"],
            "Date de publication": offre["created_at"].split("T")[0],
            "Lien": f"https://www.keejob.com{offre['absolute_url']}"
        })

# Export CSV
df = pd.DataFrame(offres)
df.to_csv("offres_keejob.csv", index=False, encoding="utf-8-sig")
print(f"‚úÖ {len(df)} offres r√©cup√©r√©es depuis l‚ÄôAPI de Keejob.")
print(df.head())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd
from bs4 import BeautifulSoup
import re

CHROMEDRIVER_PATH = "D:/telechargement/chromedriver-win64/chromedriver.exe"
service = Service(CHROMEDRIVER_PATH)
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(service=service, options=options)

def clean_text(text):
    if not text:
        return ""
    # Remplace \n, \t, etc. par espace
    text = text.replace("\n", " ").replace("\t", " ")
    # Remplace plusieurs espaces cons√©cutifs par un seul espace
    text = re.sub(r"\s+", " ", text)
    # Enl√®ve les espaces en d√©but et fin
    return text.strip()

offres = []

def extract_info(soup, label):
    try:
        b_tag = soup.find("b", string=label)
        return b_tag.find_next("br").next_sibling.strip()
    except:
        return "N/A"

for page in range(1, 80):  # Ajuste le nombre de pages si besoin
    print(f"üîé Scraping page {page}")
    driver.get(f"https://www.keejob.com/offres-emploi/?page={page}")
    time.sleep(5)

    elements = driver.find_elements(By.CSS_SELECTOR, "div.content.row-fluid")

    for el in elements:
        try:
            titre_el = el.find_element(By.CSS_SELECTOR, "div.span8 h6 a")
            titre = titre_el.text.strip()
            lien = titre_el.get_attribute("href")

            info = el.find_element(By.CSS_SELECTOR, "div.span12.no-margin-left").text.strip()
            entreprise = info.split("|")[0].strip()
            lieu_resume = info.split("\n")[-1].strip()

            # Ouvre une nouvelle fen√™tre pour le d√©tail
            driver.execute_script("window.open('');")
            driver.switch_to.window(driver.window_handles[1])
            driver.get(lien)
            time.sleep(4)

            soup = BeautifulSoup(driver.page_source, "html.parser")

            # Description compl√®te
            try:
                desc_block = soup.find("h6", string="Description de l'annonce:").find_next("div")
                description = desc_block.get_text(separator="\n", strip=True)
            except:
                description = "N/A"

            # Extraction des sections du <ol>
            try:
                ol = soup.find("ol")
                li_sections = ol.find_all("li", recursive=False)

                competences_recherchees = ""
                serait_un_plus = ""
                mission_principale = ""

                for li in li_sections:
                    titre_section = li.find("strong")
                    if titre_section:
                        titre_text = titre_section.get_text(strip=True)

                        ul = li.find("ul")
                        if ul:
                            sous_li_texts = [sli.get_text(separator=" ", strip=True) for sli in ul.find_all("li")]

                            if "Comp√©tences recherch√©es" in titre_text:
                                competences_recherchees = ", ".join(sous_li_texts)

                            elif "Serait un plus" in titre_text:
                                serait_un_plus = ", ".join(sous_li_texts)

                            elif "Mission principale" in titre_text:
                                mission_principale = ", ".join(sous_li_texts)
            except:
                competences_recherchees = "N/A"
                serait_un_plus = "N/A"
                mission_principale = "N/A"

            type_poste = extract_info(soup, "Type de poste:")
            experience = extract_info(soup, "Exp√©rience:")
            niveau_etude = extract_info(soup, "√âtude:")
            disponibilite = extract_info(soup, "Disponibilit√©:")
            langues = extract_info(soup, "Langues:")
            date_publication = extract_info(soup, "Publi√©e le:")
            lieu_travail = extract_info(soup, "Lieu de travail:")

            offres.append({
                "Titre": clean_text(titre),
                "Entreprise": clean_text(entreprise),
                "Lieu (r√©sum√© page)": clean_text(lieu_resume),
                "Lieu de travail": clean_text(lieu_travail),
                "Lien": lien,
                "Description": clean_text(description),
                "Competences_Recherchees": clean_text(competences_recherchees),
                "Serait_un_plus": clean_text(serait_un_plus),
                "Mission_principale": clean_text(mission_principale),
                "Type de poste": clean_text(type_poste),
                "Exp√©rience": clean_text(experience),
                "Niveau d'√©tude": clean_text(niveau_etude),
                "Disponibilit√©": clean_text(disponibilite),
                "Langues": clean_text(langues),
                "Date de publication": clean_text(date_publication)
            })

            driver.close()
            driver.switch_to.window(driver.window_handles[0])

        except Exception as e:
            print("‚ùå Erreur dans une offre :", e)
            continue

driver.quit()

df = pd.DataFrame(offres)
df.to_csv("offres_keejob_complet.csv", index=False, encoding="utf-8-sig")
print(f"‚úÖ Nombre total d'offres extraites : {len(df)}")
print(df.head(3))
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
import re

# ‚úÖ Ton chemin vers chromedriver
CHROMEDRIVER_PATH = "D:/telechargement/chromedriver-win64/chromedriver.exe"
service = Service(CHROMEDRIVER_PATH)
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(service=service, options=options)

offres = []

def clean_text(text):
    if text:
        return re.sub(r'\s+', ' ', text.replace('\n', ' ').replace('\t', ' ')).strip()
    return "N/A"

def extract_detail_meta(soup, label):
    try:
        metas = soup.select("div.span3.post.no-margin-left div.meta")
        for meta in metas:
            b = meta.find("b")
            if b and label.lower() in b.text.lower():
                value = b.find_next_sibling(text=True)
                return clean_text(value)
    except:
        return "N/A"

for page in range(1, 4):  # üîÅ Augmente ce chiffre pour plus de pages
    print(f"üîé Scraping page {page}")
    driver.get(f"https://www.keejob.com/offres-emploi/?page={page}")
    time.sleep(5)

    elements = driver.find_elements(By.CSS_SELECTOR, "div.content.row-fluid")
    
    for el in elements:
        try:
            titre_el = el.find_element(By.CSS_SELECTOR, "div.span8 h6 a")
            titre = titre_el.text.strip()
            lien = titre_el.get_attribute("href")

            info = el.find_element(By.CSS_SELECTOR, "div.span12.no-margin-left").text.strip()
            entreprise = info.split("|")[0].strip()
            lieu_resume = info.split("\n")[-1].strip()

            driver.execute_script("window.open('');")
            driver.switch_to.window(driver.window_handles[1])
            driver.get(lien)
            time.sleep(4)

            soup = BeautifulSoup(driver.page_source, "html.parser")

            try:
                desc_block = soup.find("h6", string=re.compile("Description de l'annonce")).find_next("div")
                description = desc_block.get_text(separator="\n", strip=True)
            except:
                description = "N/A"

            # Comp√©tences d√©taill√©es
            def extract_section(label):
                try:
                    li = soup.find("li", string=re.compile(label, re.IGNORECASE))
                    ul = li.find_next("ul")
                    return ", ".join(clean_text(li.text) for li in ul.find_all("li"))
                except:
                    return "N/A"

            competences = extract_section("Comp√©tences recherch√©es")
            serait_un_plus = extract_section("Serait un plus")
            mission_principale = extract_section("Mission principale")

            offres.append({
                "Titre": titre,
                "Entreprise": entreprise,
                "Lieu (r√©sum√© page)": lieu_resume,
                "Lieu de travail": extract_detail_meta(soup, "Lieu de travail"),
                "Lien": lien,
                "Description": description,
                "Competences_Recherchees": competences,
                "Serait_un_plus": serait_un_plus,
                "Mission_principale": mission_principale,
                "Type de poste": extract_detail_meta(soup, "Type de poste"),
                "Exp√©rience": extract_detail_meta(soup, "Exp√©rience"),
                "Niveau d'√©tude": extract_detail_meta(soup, "√âtude"),
                "Disponibilit√©": extract_detail_meta(soup, "Disponibilit√©"),
                "Langues": extract_detail_meta(soup, "Langues"),
                "Date de publication": extract_detail_meta(soup, "Publi√©e le")
            })

            driver.close()
            driver.switch_to.window(driver.window_handles[0])

        except Exception as e:
            print("‚ùå Erreur dans une offre :", e)
            continue

driver.quit()

df = pd.DataFrame(offres)
df.to_csv("offres_keejob_complet.csv", index=False, encoding="utf-8-sig")
print(f"‚úÖ Nombre total d'offres extraites : {len(df)}")
print(df.head(3))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# URL de la page listant les offres (exemple)
url_list = "https://www.keejob.com/search/?keywords=data"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

def get_offer_details(url):
    resp = requests.get(url, headers=headers)
    soup = BeautifulSoup(resp.content, "html.parser")
    
    # Titre
    titre = soup.select_one("h2.job-title").text.strip() if soup.select_one("h2.job-title") else "N/A"
    
    sidebar = soup.select_one("div.span3.post.no-margin-left")
    def get_meta_value(label):
        if not sidebar:
            return "N/A"
        el = sidebar.find("b", string=label)
        if el:
            br = el.find_next("br")
            if br and br.next_sibling:
                return br.next_sibling.strip()
        return "N/A"
    
    details = {
        "R√©f√©rence": get_meta_value("R√©f√©rence:"),
        "Publi√©e le": get_meta_value("Publi√©e le:"),
        "Type de poste": get_meta_value("Type de poste:"),
        "Lieu de travail": get_meta_value("Lieu de travail:"),
        "Exp√©rience": get_meta_value("Exp√©rience:"),
        "√âtude": get_meta_value("√âtude:"),
        "Disponibilit√©": get_meta_value("Disponibilit√©:"),
        "Langues": get_meta_value("Langues:")
    }
    
    desc_div = soup.find("div", class_="block_a span12 no-margin-left")
    description = desc_div.get_text(separator="\n").strip() if desc_div else "N/A"
    
    # Nettoyer description
    import re
    description = re.sub(r'\n\s*\n+', '\n\n', description)
    
    # Nettoyer lieux
    lieu_nettoye = ", ".join([l.strip() for l in details["Lieu de travail"].split(",") if l.strip()])
    langues_nettoye = ", ".join([l.strip() for l in details["Langues"].split(",") if l.strip()])
    
    # Construire dictionnaire complet
    result = {
        "Titre": titre,
        "R√©f√©rence": details["R√©f√©rence"],
        "Date de publication": details["Publi√©e le"],
        "Type de poste": details["Type de poste"],
        "Lieu de travail": lieu_nettoye,
        "Exp√©rience": details["Exp√©rience"],
        "Niveau d'√©tude": details["√âtude"],
        "Disponibilit√©": details["Disponibilit√©"],
        "Langues": langues_nettoye,
        "Description": description,
        "Lien": url
    }
    return result

def get_list_offers(page_url):
    resp = requests.get(page_url, headers=headers)
    soup = BeautifulSoup(resp.content, "html.parser")
    
    offers = []
    # Les offres sont dans des divs avec class "job_item"
    for job_div in soup.find_all("div", class_="job_item"):
        link_tag = job_div.find("a", href=True)
        if link_tag:
            link = "https://www.keejob.com" + link_tag['href']
            offers.append(link)
    return offers

# Exemple : r√©cup√©ration de la premi√®re page
list_offers_urls = get_list_offers(url_list)

all_offers_data = []
for i, offer_url in enumerate(list_offers_urls):
    print(f"R√©cup√©ration offre {i+1}/{len(list_offers_urls)} : {offer_url}")
    data = get_offer_details(offer_url)
    all_offers_data.append(data)
    time.sleep(1)  # Respecter le site, √©viter trop de requ√™tes trop rapides

# Sauvegarder dans DataFrame
df = pd.DataFrame(all_offers_data)
print(df.head())

# Optionnel : sauvegarder au format CSV
df.to_csv("offres_keejob.csv", index=False, encoding="utf-8-sig")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
import random

# Configuration du WebDriver
CHROMEDRIVER_PATH = "D:/telechargement/chromedriver-win64/chromedriver.exe"
service = Service(CHROMEDRIVER_PATH)
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
# options.add_argument("--headless")  # Uncomment for headless mode to reduce detection
driver = webdriver.Chrome(service=service, options=options)
wait = WebDriverWait(driver, 10)

def clean_text(text):
    if not text:
        return ""
    text = text.replace("\n", " ").replace("\t", " ")
    text = re.sub(r"\s+", " ", text)
    return text.strip()

def extract_description(soup):
    try:
        description_blocks = soup.select("div.info span")
        description = " ".join([clean_text(block.get_text()) for block in description_blocks])
        # Remove "Description de l'annonce :", "Imprimer", and "Postuler" if present
        description = re.sub(r"Description de l'annonce\s*:|Imprimer|Postuler", "", description, flags=re.I)
        return clean_text(description)
    except Exception as e:
        print(f"Erreur lors de l'extraction de la description : {e}")
        return "N/A"

def extract_offer_details(soup, bloc):
    try:
        title_tag = bloc.select_one("h2 a.titreJob")
        titre = clean_text(title_tag.text) if title_tag else "N/A"
        lien = "https://www.rekrute.com" + title_tag['href'] if title_tag and 'href' in title_tag.attrs else "N/A"

        description = extract_description(bloc)

        date_block = bloc.select_one("em.date")
        dates = clean_text(date_block.text) if date_block else "N/A"

        infos_ul = bloc.select_one("div.info ul")
        li_items = infos_ul.find_all("li") if infos_ul else []

        secteur, fonction, experience, niveau, contrat, teletravail = ["N/A"] * 6
        for li in li_items:
            txt = clean_text(li.text)
            if "Secteur d'activit√©" in txt:
                secteur = txt.split(":", 1)[-1].strip()
            elif "Fonction" in txt:
                fonction = txt.split(":", 1)[-1].strip()
            elif "Exp√©rience requise" in txt:
                experience = txt.split(":", 1)[-1].strip()
            elif "Niveau d'√©tude demand√©" in txt:
                niveau = txt.split(":", 1)[-1].strip()
            elif "Type de contrat propos√©" in txt:
                parts = txt.split("- T√©l√©travail :")
                contrat = parts[0].split(":", 1)[-1].strip()
                if len(parts) > 1:
                    teletravail = parts[1].strip()

        return {
            "Titre": titre,
            "Lien": lien,
            "Description": description,
            "Dates": dates,
            "Secteur d'activit√©": secteur,
            "Fonction": fonction,
            "Exp√©rience requise": experience,
            "Niveau d'√©tude": niveau,
            "Type de contrat": contrat,
            "T√©l√©travail": teletravail
        }
    except Exception as e:
        print(f"‚ùå Erreur d'extraction pour une offre : {e}")
        return {}

def is_relevant_for_ia_internship(titre, description):
    internship_keywords = ["stage", "internship", "pfe", "stagiaire"]
    ai_keywords = [
        "ia", "intelligence artificielle", "machine learning", "data science",
        "deep learning", "ai", "python", "big data", "analyse de donn√©es",
        "tensorflow", "pytorch", "data scientist", "data engineer"
    ]
    text = (titre + " " + description).lower()
    has_internship = any(keyword in text for keyword in internship_keywords)
    has_ai = any(keyword in text for keyword in ai_keywords)
    is_relevant = has_internship and has_ai
    if is_relevant:
        print(f"Offre pertinente trouv√©e: {titre}")
    return is_relevant

offres = []
page = 1
base_url = "https://www.rekrute.com/offres-emploi-tunisie.html"

while True:
    print(f"üîé Scraping page {page}")
    driver.get(f"{base_url}?page={page}")
    time.sleep(random.uniform(2, 4))  # D√©lai pour √©viter les blocages

    try:
        wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.col-sm-10.col-xs-12")))
        soup = BeautifulSoup(driver.page_source, "html.parser")
        blocs = soup.select("div.col-sm-10.col-xs-12")
        print(f"Nombre d'offres d√©tect√©es sur la page {page} : {len(blocs)}")

        if not blocs:
            print(f"üö´ Aucune offre trouv√©e sur la page {page}. Fin du scraping.")
            break

        for index, bloc in enumerate(blocs):
            offer_data = extract_offer_details(soup, bloc)
            if offer_data and is_relevant_for_ia_internship(offer_data["Titre"], offer_data["Description"]):
                offres.append(offer_data)
                print(f"Offre {index + 1} sur page {page}: {offer_data['Titre']} | Pertinent: True")
            else:
                print(f"Offre {index + 1} sur page {page}: {offer_data.get('Titre', 'N/A')} | Pertinent: False")

        # V√©rifier la pr√©sence du bouton "Suivant"
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, "ul.pagination li a[aria-label='Page suivante']")
            if "disabled" in next_button.get_attribute("class") or not next_button.is_enabled():
                print(f"üö´ Bouton 'Suivant' d√©sactiv√© sur la page {page}. Fin du scraping.")
                break
            next_button.click()
            page += 1
            time.sleep(random.uniform(2, 4))
        except NoSuchElementException:
            print(f"üö´ Bouton 'Suivant' non trouv√© sur la page {page}. Fin du scraping.")
            break
        except Exception as e:
            print(f"‚ùå Erreur lors du clic sur le bouton 'Suivant' sur la page {page} : {e}")
            break
    except Exception as e:
        print(f"‚ùå Erreur sur la page {page} : {e}")
        break

driver.quit()

# Sauvegarde dans un fichier CSV
df = pd.DataFrame(offres)
df.to_csv("offres_rekrute.csv", index=False, encoding="utf-8-sig")
print(f"‚úÖ Total offres extraites : {len(df)}")
print(df.head(3))
```
